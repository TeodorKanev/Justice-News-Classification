```python
import pandas as pd
```

```python
df = pd.read_csv('Data/raw_data.csv')
```

```python
def clean_atr_list(list_):
    return list_.strip('][').split(', ')
```

```python
df['Topics'] = df['Topics'].apply(clean_atr_list)
```

```python
df['Processed_Text'] = (df['Title'] + '. ' + df['Text'])
```

```python
def clean_text(text):
    return text.replace('\xa0', ' ')
```

```python
df.dropna(inplace = True)
df.reset_index(drop=True, inplace=True)
```

```python
empty_topic_indexes = df["Topics"].apply(lambda x: x == [''])
df.drop(df.loc[empty_topic_indexes].index, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df['Processed_Text'] = df['Processed_Text'].apply(clean_text)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
import re
import sys
import warnings
```

```python
#nltk.download('stopwords')
```

```python
def cleanPunc(sentence):
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    cleaned = cleaned.strip()
    cleaned = cleaned.replace("\n"," ")
    return cleaned
```

```python
df['Processed_Text'] = df['Processed_Text'].str.lower()
df['Processed_Text'] = df['Processed_Text'].apply(cleanPunc)
```

```python
def removeStopWords(sentence):
    stop_words = set(stopwords.words('english'))
    return ' '.join([word for word in sentence.split() if not word in stop_words])
```

```python
df['Processed_Text'] = df['Processed_Text'].apply(removeStopWords)
```

```python
stemmer = SnowballStemmer("english")
```

```python
def stemming(sentence):
    stemSentence = ""
    for word in sentence.split():
        stem = stemmer.stem(word)
        stemSentence += stem
        stemSentence += " "
    stemSentence = stemSentence.strip()
    return stemSentence
```

```python
df['Processed_Text'] = df['Processed_Text'].apply(stemming)
```
